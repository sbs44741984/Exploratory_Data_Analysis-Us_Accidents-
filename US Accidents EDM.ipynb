**start time**

```python
df.Start_Time[:20]
```

```python
df.rename(columns={'original_timestamp_column_name': 'Start_Time'}, inplace=True)

# Preprocess 'Start_Time' to remove unwanted trailing characters
df['Start_Time'] = df['Start_Time'].str.split('.').str[0]

# Convert 'Start_Time' to datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Check how many entries were converted to NaT
nat_count = df['Start_Time'].isna().sum()
print(f"Number of NaT entries: {nat_count}")

# Optionally, investigate NaT entries
nat_entries = df[df['Start_Time'].isna()]
print(nat_entries['Start_Time'].head(10))

# Handle NaT entries as needed, e.g., drop them or fill with a default value
df = df.dropna(subset=['Start_Time'])  # Drop rows with NaT in 'Start_Time'
```

```python
df.rename(columns={'original_timestamp_column_name': 'Start_Time'}, inplace=True)
```
